```python
import pandas as pd
```

```python
df1=pd.read_csv('Data1.csv')
```

```python
df1
```

```python
df2=pd.read_csv('Data2.csv')
```

```python
df2
```

# IF i want to merge two same row position data set then use inner or left

```python
df3 = pd.merge(df1,df2,how='inner',on='ID')
```

```python
df3
```

# IF i want to merge two different row position data set then use outer or right

```python
m=pd.merge(df1,df2,how='outer',on='ID')
```

```python
m
```

# If i want to common data then we can use indicator

```python
m=pd.merge(df1,df2,how='outer',on='ID',indicator=True)
```

```python
m
```
